```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

30만줄 -> 2860줄

```python
df = pd.read_csv("cut/1/1.csv",header=None)

#t = np.linspace(0, 10.00006, 2560)
tt = np.linspace(0,301655,2560)
re_y = signal.resample(df[1],2560)

plt.figure(figsize=(20,5))

plt.plot(df.index, df[1], 'b-')
plt.plot(tt, re_y, 'r-')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
```

30만줄 -> 1280줄

```python
df = pd.read_csv("cut/1/1.csv",header=None)

#t = np.linspace(0, 10.00006, 2560)
tt = np.linspace(0,301655,1280)
re_y = signal.resample(df[1],1280)

plt.figure(figsize=(20,5))

plt.plot(df.index, df[1], 'b-')
plt.plot(tt, re_y, 'r-')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
```

2860줄, 1280줄 비교

```python
df = pd.read_csv("cut/1/1.csv",header=None)

t = np.linspace(0, 301655, 1280)
tt = tt = np.linspace(0,301655,2560)

re_y = signal.resample(df[1],1280)
re_yy = signal.resample(df[1],2560)

plt.figure(figsize=(20,5))

plt.plot(tt, re_yy, 'r-')
plt.plot(t, re_y, 'b-')

plt.legend(['2860', '1280'], loc='best')
plt.show()
```
