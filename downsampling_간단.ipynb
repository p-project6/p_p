```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

```python
df = pd.read_csv("심전도DB/1/1.csv",header=None)

T = np.linspace(0, 10.00006, 500003)
t = np.linspace(0, 10.00006, 2560)

re_y = signal.resample(df[1],2560)

dwn = pd.DataFrame(re_y, columns=['1'])
dwn_t = pd.DataFrame(re_y, t, columns=['1'])

for i in range(2,61):
    df = pd.read_csv("심전도DB/1/{}.csv".format(i),header=None)
    dwn[i] = signal.resample(df[1],2560)
    dwn_t[i] = signal.resample(df[1],2560)
```

```python
dwn.to_csv('re1.csv') # x축 : 0 ~ 2559
dwn_t.to_csv('re1_t.csv') # x축 : 시간
```

```python
df = pd.read_csv("심전도DB/1/1.csv",header=None)

t = np.linspace(0, 10.00006, 2560)
re_y = signal.resample(df[1],2560)

plt.figure(figsize=(20,5))

plt.plot(T, df[1], 'g-')
plt.plot(t, re_y, 'r-')
plt.legend(['data', 'resampled'], loc='best')
plt.show()
```
