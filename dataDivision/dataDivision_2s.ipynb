```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
from heartpy import filter_signal
import os
import shutil
```

### *데이터합치기

```python
for j in range(1,101):
    for k in range(1,61):
        index=1
        for i in range(1,6,2):
            a = pd.read_csv("dataDivision/band/{}/{}/{}.csv".format(j,k,i), header=None)
            b= pd.read_csv("dataDivision/band/{}/{}/{}.csv".format(j,k,i+1), header=None)
            list=pd.concat([a,b])
            list.reset_index(inplace=True, drop=True)
            if os.path.exists("dataDivision/band/{}/{}".format(j,k,i+1)):
                os.makedirs('dataDivision_2/band/{}/{}'.format(j,k), exist_ok=True)
                list.to_csv('dataDivision_2/band/{}/{}/{}.csv'.format(j,k,index), index=False, header=False)
                index+=1
```

### *데이터분리/통합

- band

```python
total=pd.DataFrame()
for j in range(1,101):
    arr = np.array([])
    for l in range(180):
        arr = np.append(arr, np.array([j]))
    list=pd.DataFrame()
    for k in range(1,61):
        for i in range(1,4):
            a = pd.read_csv("dataDivision_2/band/{}/{}/{}.csv".format(j,k,i), header=None)
            a=a.transpose()
            list = list.append(a)
            
    list[512]=arr
    list.reset_index(inplace=True, drop=True)
    total=total.append(list)
total.reset_index(inplace=True, drop=True)
```
